# Import Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import plotly.graph_objects as go
import tensorflow as tf
import pandas as pd
import yfinance as yf
from datetime import datetime
```

# Import and Process Data

### a. Get Data from YFinance

```python
stock = yf.Ticker("BBCA.JK")
```

```python
stock_history = stock.history(period="10y")
```

```python
stock_history = stock_history[["Close", "High", "Low", "Volume"]]
```

### Add Moving Average

```python
stock_history["10MA"] = stock_history["Close"].rolling(window=10).mean()
stock_history["20MA"] = stock_history["Close"].rolling(window=20).mean()
```

### Add Stochastic Oscillator

```python
stock_history["K-Stochastic20"] = 100*(stock_history["Close"] - stock_history["Low"].rolling(window=20).min())/(stock_history["High"].rolling(window=20).max() - stock_history["Low"].rolling(window=20).min())
stock_history["D-Stochastic20"] = stock_history["K-Stochastic20"].rolling(window=3).mean()
```

### Delete NaN Values

```python
stock_history.dropna(inplace=True)
stock_history = stock_history.loc[(stock_history != 0).all(axis=1)]
```

### Retrieve Required Features for Prediction

```python
stock_history.drop(columns=["High", "Low"], axis=1, inplace=True)
```

```python
stock_history.head()
```

```python
data_arr = np.array(stock_history.values)
```

### Preprocess the Data

```python
def normalize_data(data, max, min) :
  new_data = data.copy()
  new_data -= min
  new_data /= (max - min)
  return new_data
```

```python
def reverse_transform(data, max, min) :
  new_data = data.copy()
  new_data *= (max - min)
  new_data += min
  return new_data
```

```python
scaled_data = normalize_data(data_arr, data_arr.max(axis=0), data_arr.min(axis=0))
```

```python
def split_data(df, size_percentage):
    """
    df : array of close price (np array)
    size : percentage number of training data (float)
    """
    size = int(size_percentage*len(df))
    train_data = df[:size]
    test_data = df[size:]
    return train_data, test_data
```

```python
stock_train, stock_test = split_data(scaled_data, 0.8)
```

```python
def windowed_dataset(data, n_past, n_future, batch_size, buff_size):
    dataset = tf.data.Dataset.from_tensor_slices(data)
    dataset = dataset.window(n_past+n_future, shift=1, drop_remainder=True)
    dataset = dataset.flat_map(lambda window: window.batch(n_past+n_future))
    dataset = dataset.map(lambda window: (window[:n_past], window[n_past:, :1]))
    dataset = dataset.shuffle(buff_size)
    dataset = dataset.batch(batch_size).prefetch(1)
    return dataset
```

```python
N_PAST = [50, 100, 200]
N_FEATURE = len(stock_history.columns)
N_FUTURE = 10
BATCH_SIZE = 64
BUFF_SIZE = 500
```

# LSTM Modelling

```python
tf.keras.backend.clear_session()
```

```python
def build_model(window, i) :
  model = tf.keras.models.Sequential([
      tf.keras.layers.Input(shape=(window, N_FEATURE)),
      tf.keras.layers.LSTM(128, return_sequences=True),
      tf.keras.layers.LSTM(128),
      tf.keras.layers.Dense(64, activation='relu'),
      tf.keras.layers.Dropout(0.2),
      tf.keras.layers.Dense(N_FUTURE)
  ], name='model_'+str(i))

  return model
```

# Training the Model

```python
def compile_model(model) :
  lr = 1e-3
  optimizer = tf.keras.optimizers.Adam(learning_rate=lr)
  model.compile(loss='mae',
                optimizer=optimizer,
                metrics=['mse'])
```

```python
class myCallback(tf.keras.callbacks.Callback) :
    def on_epoch_end(self, epoch, logs=None) :
        if logs.get('loss') <= 0.02 and abs(logs.get('loss') - logs.get('val_loss')) <= 0.01 :
            self.model.stop_training = True
```

```python
mycallback = myCallback()
```

```python
model_histories = []
models = []
for i, window in enumerate(N_PAST) :
    windowed_train = windowed_dataset(stock_train, window, N_FUTURE, BATCH_SIZE, BUFF_SIZE)
    windowed_test = windowed_dataset(stock_test, window, N_FUTURE, BATCH_SIZE, BUFF_SIZE)
    model = build_model(window, i)
    compile_model(model)
    print("\nTraining Model {}".format(i+1))
    print("----------------------------------------------------")
    history = model.fit(windowed_train, validation_data=windowed_test, epochs=100, callbacks=[mycallback])
    model_histories.append(history)
    models.append(model)
```

```python
for i, history in enumerate(model_histories) :
    print("\nModel {} : ".format(i+1))
    print("----------------------------")
    print("MAE  = {}\tVal. MAE  = {}".format(history.history["mae"][-1], history.history["val_mae"][-1]))
    print("RMSE = {}\tVal. RMSE = {}\n".format(np.sqrt(history.history["mse"][-1]), np.sqrt(history.history["val_mse"][-1])))
```

```python
model = models[0]
```

```python
N_PAST = 50
```

# Forecasting the Testing Data

```python
def model_forecast(model, series, n_past, n_future, batch_size, buff_size):
    dataset = tf.data.Dataset.from_tensor_slices(series)
    dataset = dataset.window(n_past, shift=n_future, drop_remainder=True)
    dataset = dataset.flat_map(lambda w: w.batch(n_past))
    dataset = dataset.batch(batch_size).prefetch(1)
    forecast = model.predict(dataset)
    return forecast
```

### Test Set

```python
test_size = len(stock_test)
forecast_data = scaled_data[-N_PAST - test_size:]
pred = model_forecast(model, forecast_data, N_PAST, N_FUTURE, batch_size=BATCH_SIZE, buff_size=BUFF_SIZE)
```

```python
pred = np.reshape(pred, (-1,))
```

```python
pred = pred[:test_size]
```

```python
stock_pred = reverse_transform(pred, data_arr[:, 0].max(), data_arr[:, 0].min())
```

```python
plt.figure(figsize=(10,8))
plt.plot(np.arange(test_size), data_arr[-test_size:, 0], label="True")
plt.plot(np.arange(test_size), stock_pred, label="Forecast")
plt.legend()
plt.show()
```

```python
from sklearn.metrics import mean_absolute_error
```

```python
test_mae = mean_absolute_error(pred, scaled_data[-test_size:, 0])
test_mae
```

### Train Set

```python
train_data = scaled_data[:-test_size]
train_pred = model_forecast(model, train_data, N_PAST, N_FUTURE, batch_size=BATCH_SIZE, buff_size=BUFF_SIZE)
```

```python
train_pred = np.reshape(train_pred, (-1,))
```

```python
train_size = len(scaled_data[N_PAST:-test_size])
train_size
```

```python
train_pred = train_pred[:train_size]
```

```python
train_pred_rev = reverse_transform(train_pred, data_arr[:, 0].max(), data_arr[:, 0].min())
```

```python
plt.figure(figsize=(10,8))
plt.plot(np.arange(train_size), data_arr[N_PAST:-test_size, 0], label="True")
plt.plot(np.arange(train_size), train_pred_rev, label="Forecast")
plt.legend()
plt.show()
```

```python
train_mae = mean_absolute_error(train_pred, scaled_data[N_PAST:-test_size, 0])
train_mae
```
